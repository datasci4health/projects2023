### Análise PCA e Clustering

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
file_location = '../data/processed/dados-tratados.xlsx'
data = pd.read_excel(file_location)
data['pre-infection_category'] = df['pre-infection_category']
```

```python
data
```

```python
# cópia dos dados para normalização
data_copy = data.copy()

# definir variáveis preditoras sem a variável alvo
predictors = data.columns.tolist()
predictors.remove('max_dg_sev_during_hospitalization')

# normalização dos dados
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_copy[predictors])

# aplicar o PCA
pca = PCA()
pca.fit(data_scaled)

# visualizar a variância de cada componente
variances = pca.explained_variance_ratio_
print(variances)

# plot das variâncias de cada componente
plt.figure(figsize = (6, 4))
plt.plot(range(len(variances)), variances.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explained Variance by Component')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
# são necessários 9 componentes para explicar 80% da variância
# soma dos 9 primeiros componentes
print(variances[0:10].sum())
```

```python
# fazer o PCA com 9 componentes
pca = PCA(n_components = 9)
pca.fit(data_scaled)

# calcular os scores dos componentes para os elementos do dataset
scores_pca = pca.transform(data_scaled)
print(scores_pca)
```

```python
# integrar a análise de PCA com uma clusterização por K-means
wcss = []
for i in range(1, 21):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (6, 4))
plt.plot(range(1, 21), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('K-means with PCA Clustering')
plt.show()
```

```python
# I will choose 4 clusters
kmeans_pca = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
kmeans_pca.fit(scores_pca)
```

```python
# I will add PCA scores and assigned clusters to the database
data_pca_kmeans = pd.concat([data.reset_index(drop = True), pd.DataFrame(scores_pca)], axis = 1)
data_pca_kmeans.columns.values[-9:] = ['component1', 'component2', 'component3', 'component4',
                                       'component5', 'component6', 'component7', 'component7',
                                       'component9']
data_pca_kmeans['label'] = kmeans_pca.labels_
# data_pca_kmeans['cluster'] = data_pca_kmeans['label'].map({0: 'first', 1: 'second',
#                                                         2: 'third', 3: 'fourth'})
```

```python
 import plotly.graph_objects as go
 
 def plot_cluster(cluster_data):
    age_labels = {1: '<50', 2: '50-64', 3: '>64'}
    temp_df = cluster_data[['age_group', 'gender']].copy()
    temp_df['age_group'] = temp_df['age_group'].map(age_labels)

    grouped_counts = temp_df.groupby(['age_group', 'gender']).size().reset_index(name='Count')

    fig = go.Figure(data=[
        go.Bar(name='Masculino', x=grouped_counts[grouped_counts['gender'] == 0]['age_group'], y=grouped_counts[grouped_counts['gender'] == 0]['Count']),
        go.Bar(name='Femenino', x=grouped_counts[grouped_counts['gender'] == 1]['age_group'], y=grouped_counts[grouped_counts['gender'] == 1]['Count'])
    ])

    fig.update_layout(
        title= 'Distribuição dos pacientes por grupo de idade e gênero',
        xaxis=dict(title='Faixa etária'),
        yaxis=dict(title='Quantidade'),
        barmode='group',
        legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='center', x=1)
    )
    fig.show()


    severity_labels = {0: 'Mild', 1: 'Moderate', 2: 'Severe', 3: 'Critical'}
    temp_df = cluster_data[['max_dg_sev_during_hospitalization']].copy()
    severity_colors = {'Mild': '#00bfff', 'Moderate': '#ffa500', 'Severe': '#ff0000', 'Critical': '#633974'}
    temp_df['max_dg_sev_during_hospitalization'] = temp_df['max_dg_sev_during_hospitalization'].map(severity_labels)

    severity_counts = temp_df['max_dg_sev_during_hospitalization'].value_counts().reset_index().rename(columns={'max_dg_sev_during_hospitalization': 'Severity'})

    fig = go.Figure(data=[
        go.Bar(x=severity_counts['Severity'], y=severity_counts['count'], marker_color=[severity_colors[severity] for severity in severity_counts['Severity']])
    ])

    fig.update_layout(
        title='Gravidade máxima do diagnóstico durante a hospitalização',
        xaxis=dict(title='Gravidade'),
        yaxis=dict(title='Quantidade'),
    )

    fig.show()


    BMI_labels = {0: 'Não-obeso', 1: 'Obeso'}
    temp_df = cluster_data[['BMI']].copy()
    BMI_colors = {'Obeso': '#00bfff', 'Não-obeso': '#ffa500'}
    temp_df['BMI'] = temp_df['BMI'].map(BMI_labels)

    BMI_counts = temp_df['BMI'].value_counts().reset_index().rename(columns={'BMI': 'BMI'})

    fig = go.Figure(data=[
        go.Bar(x=BMI_counts['BMI'], y=BMI_counts['count'], marker_color=[BMI_colors[BMI] for BMI in BMI_counts['BMI']])
    ])

    fig.update_layout(
        title='Classificação de IMC',
        xaxis=dict(title='IMC'),
        yaxis=dict(title='Quantidade'),
    )

    fig.show()


    # cópia do dataframe
    temp_df = cluster_data.copy()

    # Mapear as categorias de pré-infecção
    pre_infection_labels = {1: '<=19.99', 2: '20-29.99', 3: '30-39.99', 4: '40+'}
    temp_df['pre-infection_category'] = temp_df['pre-infection_category'].map(pre_infection_labels)

    # Mapear as categorias de morte ou não-morte
    death_labels = {0: 'Não', 1: 'Sim'}
    temp_df['death_during_hospitalization'] = temp_df['death_during_hospitalization'].map(death_labels)

    # Agrupar os dados
    grouped_counts = temp_df.groupby(['pre-infection_category', 'death_during_hospitalization']).size().reset_index(name='Count')

    # Criar barplot
    fig = go.Figure()

    for death in death_labels.values():
        death_data = grouped_counts[grouped_counts['death_during_hospitalization'] == death]
        fig.add_trace(go.Bar(name=death, x=death_data['pre-infection_category'], y=death_data['Count']))

    fig.update_layout(
        title='Relação entre nível de vitamina D pré-infecção e casos de morte',
        xaxis=dict(title='Nível de vitamina D pré-infecção'),
        yaxis=dict(title='Quantidade'),
        barmode='group',
        legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='center', x=1)
    )

    # Mostrar el gráfico
    fig.show()
```
