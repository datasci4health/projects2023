```python
import pandas as pd
import plotly.express as px
```

```python
lDf = pd.read_csv("database.csv")
```

```python
lDf.head()
```

```python
lDfDeaths = lDf.drop(columns=["Entity", "Code", "Year"])
lDfDeathsCause = lDfDeaths.idxmax(axis=1)
lDf["Top Cause"] = lDfDeathsCause
lDf.head()
```

```python
fig = px.choropleth(lDf,               
              locations="Code",               
              color="Top Cause",
              hover_name="Entity",  
              animation_frame="Year",
              width=1000     
)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
