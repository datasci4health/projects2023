```python
def plot_1(col, titulo):
    
    legenda = {
    0 : 'Nenhum',
    1 : 'Pouco',
    2 : 'Moderado',
    3 : 'Severo',
	4 : 'Extremo',
	5 : 'NA'}

    aux = df.query(f"{col} != -1") 
    plt.hist(aux[col])
    plt.xlabel("Resposta")
    plt.ylabel("Quantidade de cada resposta")
    plt.title(f"{titulo}")
    plt.xticks(range(0,6), [legenda[val] for val in range(0,6)])

    plt.savefig("../assets/hist_"+ col+".jpg")
    plt.show()
```

```python
plot_1('Mobility', "Quantidade de pessoas que possuem dificuldade para andar")
plot_1('Care', "Quantidade de pessoas que possuem dificuldade para se cuidar")
plot_1('Active', "Quantidade de pessoas que possuem dificuldade para atividades diarias")
plot_1('Pain', "Quantidade de pessoas que possuem dor")
plot_1('Anxious', "Quantidade de pessoas que possuem ansiedade/depressao")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Aplicar a clusterizaÃ§Ã£o
kmeans = KMeans(n_clusters=3)  # Defina o nÃºmero de clusters desejado
clusters = kmeans.fit_predict(df)

# Adicionar as informaÃ§Ãµes de cluster ao dataframe original
df['Cluster'] = clusters

# Visualizar o resultado
print(df.head())
```
