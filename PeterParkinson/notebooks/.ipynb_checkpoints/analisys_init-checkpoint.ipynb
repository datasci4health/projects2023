#### Para analisar idade dos participantes, foi agrupado por faixa etária (pegamos como referencia os quartis):

. 18 - 30 (grupo 0)

. 31 - 57 (grupo 1)

. 58 - 66 (grupo 2)

. 67 - 72 (grupo 3)

. > 72 (grupo 4)

```python
df2['age'][26]
```

```python
tam = len(df2)
age_range = []
for i in range(tam):
    #print(df2['age'][i], i)
    if df2['age'][i] <= 30:
        aux = 0
        age_range.append(aux)
        
    elif df2['age'][i] > 30 and df2['age'][i] <= 57:
        aux = 1
        age_range.append(aux)
        
    elif df2['age'][i] > 57 and df2['age'][i] <= 66:
        aux = 2
        age_range.append(aux)
        
    elif df2['age'][i] > 66 and df2['age'][i] <= 72:
        aux = 3
        age_range.append(aux)
        
    elif df2['age'][i] > 72:
        aux = 4
        age_range.append(aux)
    else:
        print("Julia deu erro aqui")
```

```python
df2['age_range'] = age_range
df2.drop(['age'], axis=1, inplace=True)
```

```python
df2
```

```python
df2.groupby(['age_range']).age_range.count().sort_values()#.plot(kind='bar')
```

```python
df2['age_range'].unique()
```

```python
import seaborn as sns

plot = sns.barplot(data=df2, x = df2['age_range'].unique(), y = df2['age_range'].value_counts(),
                  palette = "Set2")
plot.set_title( "Contagem da idade")
plot.set_xlabel( "Range de idades")
plot.set_ylabel( "Quantidade de pessoas")
```
