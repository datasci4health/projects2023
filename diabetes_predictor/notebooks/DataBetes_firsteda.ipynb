```python
limiar = 0.3
nr = df_selec_corr.shape[0]
C = df_selec_corr.to_numpy()
M = np.zeros((nr,nr))
for i in range(nr):
  for j in range(nr):
    if C[i,j] >= limiar:
      if i != j:
        M[i,j] = 1
    elif C[i,j] <= -limiar:
      if i != j:
        M[i,j] = 1
# M_p = (df_selec_corr[:]>=limiar).astype(int)
# M_n = (df_selec_corr[:]<=-limiar).astype(int)
np.set_printoptions(threshold=np.inf)
# print(M)

n = M.shape[0]
plt.figure(figsize=(8,8))
plt.rcParams.update({'font.size': 8})
plt.axis('off')
axis = np.linspace(0, 2*np.pi, n, endpoint=False)
x, y = np.cos(axis), np.sin(axis)
for i in range(n):
    for j in range(i + 1, n):
        if M[i,j] == 1:
            plt.plot((x[i], x[j]), (y[i], y[j]), color = 'b')
plt.scatter(x, y)
for i, txt in enumerate(df_selec_corr.columns.values):
    plt.annotate(txt, (x[i], y[i]))
plt.show()
```

```python
# renomeia cabeçalho 'ordem'
df2015.rename(columns = {'ordem':'chave'}, inplace = True)
df2016.rename(columns = {'ordem':'chave'}, inplace = True)

# variáveis selecionadas:
var_selec = ['chave',
             'replica',
             'ano',
             'cidade',
             'q7',    # sexo
             'q6',    # idade (anos)
             'fet',   # faixa etária (anos)
             'q8a',   # grau escolaridade
             'q75',   # pressão alta
             'q35',   # bebida alcoólica
             'q36',   # freq alcool
             'q42',   # exercício físico
             'q69',   # cor
             'q9',    # peso (kg)
             'q11',   # altura (cm)
             'r202',  # idade do ínicio do diabetes

             'q16',   # freq_horta
             'q29', # freq_refrigerante
             'moradores', # moradores
             'q60', # fumante
            #  'r900', # bolsa_familia
             'q88', # posse_plano_saude

             'q76',   # diabetes
             ]


# anos selecionados:
df_selec = pd.concat([df2015[var_selec].copy(),
                      df2016[var_selec].copy(),
                      df2017[var_selec].copy(),
                      df2018[var_selec].copy(),
                      df2019[var_selec].copy(),
                      df2020[var_selec].copy(),
                      df2021[var_selec].copy(),])

# renomeia os cabeçalhos
df_selec.rename(columns = {'q7':'sexo','q6':'idade (anos)','fet':'faixa etária (anos)','q8a':'grau escolaridade','q75':'pressão alta',
                           'q35':'bebida alcoólica','q36':'freq alcool','q42':'exercício físico','q69':'cor','q9':'peso (kg)',
                           'q11':'altura (cm)','r202':'idade início diabetes','q76':'diabetes',
                           'q16':'freq horta','q29':'freq refrigerante','q60':'fumante','r900':'bolsa família','q88':'posse plano saúde'
                           }, inplace = True)

df_selec.loc[df_selec['altura (cm)'] == 777, 'altura (cm)'] = pd.np.nan
df_selec.loc[df_selec['altura (cm)'] == 888, 'altura (cm)'] = pd.np.nan
df_selec.loc[df_selec['peso (kg)'] == 777, 'peso (kg)'] = pd.np.nan
df_selec.loc[df_selec['peso (kg)'] == 888, 'peso (kg)'] = pd.np.nan

# calcula e adiciona IMC ao dataframe
imc = df_selec['peso (kg)'] / (((df_selec['altura (cm)'])/100)**2)  # 
df_selec['imc'] = imc

def get_bmi_class(imc):
    if imc < 18.5:
        return 'baixo peso'
    elif imc < 25:
        return 'adequado ou eutrófico'
    elif imc < 30:
        return 'sobrepeso'
    elif imc < 40:
        return 'obesidade classe I'
    elif imc < 50:
        return 'obesidade classe II'
    else:
        return 'obesidade classe III'

df_selec['classe imc'] = df_selec['imc'].apply(get_bmi_class)

# Substitui as cidades
replacements = {1:'Aracajú',2:'Belém',3:'Belo Horizonte',4:'Boa Vista',5:'Campo Grande',6:'Cuibá',7:'Curitiba',8:'Florianópolis',9:'Fortaleza',10:'Goiânia',11:'João Pessoa',12:'Macapá',
                13:'Maceió',14:'Manaus',15:'Natal',16:'Palmas',17:'Porto Alegre',18:'Porto Velho',19:'Recife',20:'Rio Branco',21:'Rio de Janeiro',22:'Salvador',23:'São Luís',24:'São Paulo',
                25:'Teresina',26:'Vitória',27:'Distrito Federal',}
df_selec['cidade'] = df_selec['cidade'].replace(replacements)

# Adiciona as regiões
replacements = {'Aracajú':'Nordeste','Belém':'Norte','Belo Horizonte':'Sudeste','Boa Vista':'Norte','Campo Grande':'Centro-Oeste','Cuibá':'Centro-Oeste','Curitiba':'Sul','Florianópolis':'Sul',
                'Fortaleza':'Nordeste','Goiânia':'Centro-Oeste','João Pessoa':'Nordeste','Macapá':'Norte','Maceió':'Nordeste','Manaus':'Norte','Natal':'Nordeste','Palmas':'Norte',
                'Porto Alegre':'Sul','Porto Velho':'Norte','Recife':'Nordeste','Rio Branco':'Norte','Rio de Janeiro':'Sudeste','Salvador':'Nordeste','São Luís':'Nordeste','São Paulo':'Sudeste',
                'Teresina':'Nordeste','Vitória':'Sudeste','Distrito Federal':'Centro-Oeste',}
df_selec['região'] = df_selec['cidade'].replace(replacements)

# ordena os cabeçalhos
df_selec=df_selec[['chave',
                  'replica',
                  'ano',
                  'cidade',
                  'região',
                  'sexo',
                  'idade (anos)',
                  'faixa etária (anos)',
                  'grau escolaridade',
                  'pressão alta',
                  'bebida alcoólica',
                  'freq alcool',
                  'exercício físico',
                  'cor',
                  'peso (kg)',
                  'altura (cm)',
                  'imc',
                  'classe imc',
                  'freq horta',
                  'freq refrigerante',
                  'moradores',
                  'fumante',
                  # 'bolsa família',
                  'posse plano saúde',
                  'idade início diabetes',
                  'diabetes',
                  ]]

df_selec
```

```python
X = X.dropna() # drop rows with NaN values
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
'''
Hyperparameter Tuning – `Cross-Validation`
'''

# Creating a dictionary of parameters to use in GridSearchCV
from sklearn.model_selection import GridSearchCV

params = {
    'criterion':  ['gini', 'entropy'],
    'max_depth':  [None, 2, 4, 6, 8, 10],
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8],
    'splitter': ['best', 'random']
}

clf = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1,
)

clf.fit(X_train, y_train)
print(clf.best_params_)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(max_depth=6, 
                             criterion='entropy', 
                             max_features=None, 
                             splitter='random')

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
